# Introduction 

### Lunchinga New Bank Branch in Toronto

 A famous Canadian bank manager decided to increase the number of its branch in Toronto. Our mission is to recommend the best neighborhoods for the inauguration of this new bank branch. 

 This problem could be solved by different strategies, but in this project, we will create an indicator to prioritize the best 5 neighborhoods for Launching a new branch of the bank.

 For this purpose our indicator  is defined as follows: 
 

  #### Locating Index= (Population of the Neighborhood)* (Average Income of People living in neighborhood)/(Normalized Bank Availability Index+1)
  
  
 
 The more the "Locating Index" means a better place to lunch a new bank branch.   

Hint: the +1 on the denominator of the locationg index is defined to avoid 0 result in the denominator of the locating index.
