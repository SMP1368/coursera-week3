# Lunchinga New Bank Branch in Toronto

# Introduction

 A famous Canadian bank manager decided to increase the number of its branch in Toronto. Our mission is to recommend the best neighborhoods for the inauguration of this new bank branch. 

 This problem could be solved by different strategies, but in this project, we will create an indicator to prioritize the best 5 neighborhoods for Launching a new branch of the bank.

 For this purpose our indicator  is defined as follows: 
 

  #### Locating Index= (Population of the Neighborhood)* (Average Income of People living in neighborhood)/(Normalized Bank Availability Index+1)/10^8
  
  
 
 The more the "Locating Index" means a better place to lunch a new bank branch.

```python
import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe
```

```python
results = requests.get(url).json()
results
```

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
venues = results['response']['groups'][0]['items']
    
nearby_venues = json_normalize(venues) # flatten JSON

# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]

# filter the category for each row
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)

# clean columns
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]

nearby_venues
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
    # return only relevant information for each nearby venue
    
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
TORONTO_venues = getNearbyVenues(names=dfs_Toronto['Neighbourhood'],
                                   latitudes=dfs_Toronto['Latitude'],
                                   longitudes=dfs_Toronto['Longitude']
                                  )
```

```python
# one hot encoding
Toronto_onehot = pd.get_dummies(TORONTO_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
Toronto_onehot['Neighborhood'] = TORONTO_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [Toronto_onehot.columns[-1]] + list(Toronto_onehot.columns[:-1])
Toronto_onehot = Toronto_onehot[fixed_columns]

Toronto_onehot.head()
```

Hint: As we just need bank availibilty index, we would filter the above table by "Bank" column.

```python
Toronto_grouped = Toronto_onehot.groupby('Neighborhood').mean().reset_index()
Toronto_grouped_Bank=Toronto_grouped[['Neighborhood','Bank']]
Toronto_grouped_Bank
```

Then we calculate "Bank index" by adding 1 to "Bank" column. The aim of this process is to avoid infinity result in the case of zero "Bank" number.  

```python
Toronto_grouped_Bank['Bank index']=Toronto_grouped_Bank.Bank+1
Toronto_grouped_Bank
```

Now it is timme to merge to tables to calcalate locating index. 

```python
merge_result=pd.merge(Toronto_grouped_Bank,dfs3, left_on='Neighborhood', right_on='Name', how='inner')
merge_result
```

Then we would like to filter our necessary column for better calculation. 

```python
select_column=merge_result[['Neighborhood','Bank index','Population','Average Income']]
select_column
```

In this section we calculate Lacating index and then sort values. as mentioned before " the more the locating index, the better place for lunching new bank branch"

```python
select_column['Locating Index']= ((select_column.Population)*(select_column['Average Income']))/(select_column['Bank index'])/100000000
priority_table=select_column.sort_values(by=['Locating Index'], ascending=True)
priority_table
```

```python
merge_result2=pd.merge(dfs_Toronto,priority_table, left_on='Neighbourhood', right_on='Neighborhood', how='right')
merge_result2
```

```python
select_column2=merge_result2[['Neighborhood','Locating Index','Latitude','Longitude']]
select_column3=select_column2.sort_values(by=['Locating Index'], ascending=False)
select_column3
```
